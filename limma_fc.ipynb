# Load libraries

```r
library(limma)
library(leukemiasEset) # Example data set
```

# Prepare example data

This step is not needed when performing an analysis for other data

```r
data(leukemiasEset)
```

Select two groups for fold change analysis

```r
example_data <- leukemiasEset[, leukemiasEset$LeukemiaType %in% c("ALL", "NoL")]
```

Update data that we have only two groups now

```r
example_data$LeukemiaType <- factor(example_data$LeukemiaType)
```

# Create design matrix

This matrix tells limma which samples belong to which group.


Rownames have to be numbered, this number corresponds to the column\
first column has to be named '(Intercept)',\
Name of the second column is irrelevant,\
0: group1\
1: group2

```r
design <- model.matrix(~ example_data$LeukemiaType)
design
```

# Data matrix

!!! DATA NEEDS TO BE IN LOG2 FORMAT !!!

This is already the case for the example data, real data needs to be transformed.

```r
example_matrix <- exprs(example_data)
```

```r
head(example_matrix)
```

# FC calculation

```r
fit <- lmFit(example_matrix, design)
fit <- eBayes(fit)
```

Plot logFC sorted by p-value.\
Set number to limit number of features printed, or set to very high number to print all.\
Set p.value to print only significant.

```r
topTable(fit, number = 99999, p.value = 0.05)
```
